```python
import math
from qiskit import QuantumCircuit

numeri = input('inserisci la lista dei coeff : ')
lista_coeff = [float(x) for x in numeri.split()]
print(lista_coeff)

norma = (sum([x**2 for x in lista_coeff]))**(1/2) # somma dei quadrati (alfa in qiskit)
# abbiamo un vettore di N elementi, 
# vogliamo scrivere il vettore come comb lin di N vettori di base i cui coeff sono le componenti del vettore
# quindi vogliamo costruire un circuito con n = math.ceil(log2(N)) bit 
N = len(lista_coeff)
n = math.ceil(math.log2(N))
# riempio le componenti fino a 2^N con zeri se serve (cioè se N non è una potenza di 2)
for i in range(2**n-N,2**n) :
    lista_coeff.append(0)
# desired_state è |phi> che rappresenta sigma_0 = lista_coeff
# (es : lista_coeff=[4,8,5] e desired_state = [4/alfa,8/alfa,5/alfa,0] (abbiamo aggiunto 0 prima) )
desired_state = [x/norma for x in lista_coeff]
print(desired_state)

qc = QuantumCircuit(n)
# inizializza il circuito a sigma_0
#qc.initialize(desired_state,range(n)) # secondo input : lista dei pedici dei qubit nel circuito
#qc.decompose(reps=8).draw(output="mpl")
```
